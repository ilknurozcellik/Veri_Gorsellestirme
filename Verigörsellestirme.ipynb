```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veri setini yükleme
tips = sns.load_dataset("tips")

# 1. Toplam Hesap (total_bill) ve Bahşiş (tip) Arasındaki İlişki
plt.figure(figsize=(8,6)) # yeni bir grafik oluşturulur. oluşturulan grafiğin genişliği 8, yüksekliği 6 dır.
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", alpha=0.7) # scatterplot dağılım grafiği oluşturuyor.
plt.title("Toplam Hesap vs Bahşiş") # grafiğin başlığı 
plt.xlabel("Toplam Hesap (total_bill)") 
plt.ylabel("Bahşiş (tip)")
plt.show() # oluşturduğumuz grafikleri gözlemleyebilmemiz için show fonksiyonunu kullanma zorundayız.


# 2. Günlere Göre Toplam Hesap Dağılımı (Boxplot)
plt.figure(figsize=(8,6)) 
sns.boxplot(data=tips, x="day", y="total_bill", hue="day", palette="Set2", legend=False) # boxplot kullanılarak kutu grafiği oluşturulur.
plt.title("Günlere Göre Toplam Hesap Dağılımı") # grafiğin başlığı  
plt.xlabel("Gün")  # gün etiketi x eksenine, toplam hesap etiketi y eksenine yazılır.
plt.ylabel("Toplam Hesap (total_bill)")
plt.show()

# 3. Masa Büyüklüğüne Göre Bahşiş Oranları (Barplot)
tips["tip_rate"] = (tips["tip"] / tips["total_bill"]) * 100
plt.figure(figsize=(8,6))
sns.barplot(data=tips, x="size", y="tip_rate", hue="size", palette="viridis", legend=False)
plt.title("Masa Büyüklüğüne Göre Bahşiş Oranı")
plt.xlabel("Masa Büyüklüğü (size)")
plt.ylabel("Ortalama Bahşiş Oranı (%)")
plt.show()

# 4. Günün Zamanına Göre Bahşiş Dağılımı (KDE Plot)
plt.figure(figsize=(8,6))
sns.kdeplot(data=tips[tips["time"] == "Lunch"], x="tip", fill=True, label="Lunch", alpha=0.5)  # öğle yemeğinde verilen bahşiş verilerini alır.
sns.kdeplot(data=tips[tips["time"] == "Dinner"], x="tip", fill=True, label="Dinner", alpha=0.5) # akşam yemeğinde verilen bahşiş verisini alır.
plt.title("Öğle ve Akşam Bahşiş Dağılımı")
plt.xlabel("Bahşiş (tip)")
plt.ylabel("Yoğunluk")
plt.legend() # grafik üzerinde açıklama eklememizi sağlıyor 
plt.show()


# 5. Korelasyon Analizi (Heatmap)
plt.figure(figsize=(8,6))
sns.heatmap(tips.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
# tips.select_dtypes(include=["number"]): sadece sayısal sütunları seçebilmemiz için bu kodu kullanmalıyız.
plt.title("Korelasyon Matrisi") # başlık
plt.show()
```
