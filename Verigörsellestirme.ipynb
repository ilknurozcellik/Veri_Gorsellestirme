```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veri setini ykleme
tips = sns.load_dataset("tips")

# 1. Toplam Hesap (total_bill) ve Bahi (tip) Arasndaki liki
plt.figure(figsize=(8,6)) # yeni bir grafik oluturulur. oluturulan grafiin genilii 8, ykseklii 6 dr.
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", alpha=0.7) # scatterplot dalm grafii oluturuyor.
plt.title("Toplam Hesap vs Bahi") # grafiin bal 
plt.xlabel("Toplam Hesap (total_bill)") 
plt.ylabel("Bahi (tip)")
plt.show() # oluturduumuz grafikleri gzlemleyebilmemiz iin show fonksiyonunu kullanma zorundayz.


# 2. Gnlere Gre Toplam Hesap Dalm (Boxplot)
plt.figure(figsize=(8,6)) 
sns.boxplot(data=tips, x="day", y="total_bill", hue="day", palette="Set2", legend=False) # boxplot kullanlarak kutu grafii oluturulur.
plt.title("Gnlere Gre Toplam Hesap Dalm") # grafiin bal  
plt.xlabel("Gn")  # gn etiketi x eksenine, toplam hesap etiketi y eksenine yazlr.
plt.ylabel("Toplam Hesap (total_bill)")
plt.show()

# 3. Masa Byklne Gre Bahi Oranlar (Barplot)
tips["tip_rate"] = (tips["tip"] / tips["total_bill"]) * 100
plt.figure(figsize=(8,6))
sns.barplot(data=tips, x="size", y="tip_rate", hue="size", palette="viridis", legend=False)
plt.title("Masa Byklne Gre Bahi Oran")
plt.xlabel("Masa Bykl (size)")
plt.ylabel("Ortalama Bahi Oran (%)")
plt.show()

# 4. Gnn Zamanna Gre Bahi Dalm (KDE Plot)
plt.figure(figsize=(8,6))
sns.kdeplot(data=tips[tips["time"] == "Lunch"], x="tip", fill=True, label="Lunch", alpha=0.5)  # le yemeinde verilen bahi verilerini alr.
sns.kdeplot(data=tips[tips["time"] == "Dinner"], x="tip", fill=True, label="Dinner", alpha=0.5) # akam yemeinde verilen bahi verisini alr.
plt.title("le ve Akam Bahi Dalm")
plt.xlabel("Bahi (tip)")
plt.ylabel("Younluk")
plt.legend() # grafik zerinde aklama eklememizi salyor 
plt.show()


# 5. Korelasyon Analizi (Heatmap)
plt.figure(figsize=(8,6))
sns.heatmap(tips.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
# tips.select_dtypes(include=["number"]): sadece saysal stunlar seebilmemiz iin bu kodu kullanmalyz.
plt.title("Korelasyon Matrisi") # balk
plt.show()
```
